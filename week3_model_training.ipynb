```python
plt.figure(figsize=(10, 6))
top = importance.head(15).iloc[::-1]
plt.barh(top['feature'], top['importance'], color='#7aa7ff')
plt.title('Week 3 Feature Importance')
plt.xlabel('Importance')
plt.tight_layout()
plt.show()
```

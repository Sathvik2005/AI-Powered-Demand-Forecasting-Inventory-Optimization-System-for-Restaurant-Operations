# Week 4: Evaluation, Feature Importance, and Business Reporting

This notebook translates the Week 3 model outputs into business metrics, feature explanations, and ready-to-share visuals.

```python
!pip install -q kagglehub scikit-learn xgboost matplotlib seaborn

from week3_week4_suite.shared import build_dashboard_payload, generate_artifacts
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('dark_background')
```

```python
payload = build_dashboard_payload()
artifacts = generate_artifacts()
forecast = pd.DataFrame(payload['forecast'])
comparison = pd.DataFrame(payload['comparison'])
insights = payload['insights']

print('Best model:', payload['best_model_name'])
print('MAE:', round(payload['best_metrics']['mae'], 2))
print('RMSE:', round(payload['best_metrics']['rmse'], 2))
print('Artifacts:', artifacts)
```

```python
plt.figure(figsize=(14, 6))
plt.plot(forecast['date'], forecast['actual'], label='Actual', linewidth=2)
plt.plot(forecast['date'], forecast['predicted'], label='Predicted', linewidth=2)
plt.title('Week 4 Forecast Overlay')
plt.xlabel('Date')
plt.ylabel('Orders')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
top = pd.DataFrame(payload['feature_importance']).head(15).iloc[::-1]
plt.barh(top['feature'], top['importance'], color='#78f0c4')
plt.title('Week 4 Feature Importance')
plt.xlabel('Importance')
plt.tight_layout()
plt.show()
```

```python
print('Business summary')
print('Average daily demand:', round(insights['avg_daily_demand'], 2))
print('Weekend uplift (%):', round(insights['weekend_uplift_pct'], 2))
print('Detected anomalies:', insights['anomaly_days'])
print('Top weekday index:', insights['top_weekday'])

comparison
```
